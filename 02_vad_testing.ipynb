```python
import json
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "2"
with open("secrets.json") as f:
    secrets = json.load(f)
secrets.keys()
import torch
```

```python
from pyannote.audio import Model
model = Model.from_pretrained("pyannote/segmentation", 
                              use_auth_token=secrets["hf_token"]).to(torch.device("cuda"))

from pyannote.audio.pipelines import VoiceActivityDetection
pipeline = VoiceActivityDetection(segmentation=model)
HYPER_PARAMETERS = {
  # onset/offset activation thresholds
  "onset": 0.5, "offset": 0.5,
  # remove speech regions shorter than that many seconds.
  "min_duration_on": 0.1,
  # fill non-speech regions shorter than that many seconds.
  "min_duration_off": 0.5,
}
pipeline.instantiate(HYPER_PARAMETERS, )


output = pipeline("data/audio_16khz_mono_wav/0.wav")
```

```python
str_to_write = output.to_rttm()
from pathlib import Path
Path("data/vad/").mkdir(exist_ok=True)
outpath = Path("data/vad/0.rttm")
if outpath.exists():
    outpath.unlink()
outpath.write_text(str_to_write)
```
